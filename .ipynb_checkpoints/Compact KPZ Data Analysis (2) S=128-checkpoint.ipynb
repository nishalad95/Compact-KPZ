Calibration of KPZ - set lambda = 0.0, Lattice size S = 128

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
avgEnergy = [0] * 21
avgNumVortices = [0] * 21
avgEnergySquared = [0] * 21
CL = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0] # 'temperature', pump power
CL.insert(7, 3.2)
CL.insert(8, 3.3)
CL.insert(9, 3.4)
CL.insert(11, 3.6)
CL.insert(12, 3.7)
CL.insert(13, 3.8)
print(CL)
```

```python
def PlotGraphs(r, cl, N, cl_index, noise=True):
    R = r
    fig = plt.figure(figsize=(5, 3))
    if noise:
        stocAvg = 0.0
        stocAvg2 = 0.0
        stocVortex = 0.0
    for i in range(1,R+1):
        energy = np.loadtxt("Data_S=128/CL_" + str(cl) + "/Energy_CL" + str(cl) + "_N" + str(N) + "_R" + str(i) + ".txt", dtype=float)
        iteration = np.arange(0, len(energy), 1)
        plt.plot(iteration, energy)
        if noise:
            stocAvg += energy[-1]
            stocAvg2 += energy[-1]**2
    if noise:
        stocAvg2 /= R
        stocAvg /= R
    plt.title("Energy as a function of time for a $128$x$128$ matrix, \n over " + str(R) + "\
    realizations, $C_l =" + str(cl) + "$, $N =" + str(N) + "$, $\lambda = 0$")
    plt.xlabel("time [T]")
    plt.ylabel("E")
    plt.show()

    fig = plt.figure(figsize=(5, 3))
    plt.xlabel("time [T]")
    plt.ylabel("Number of vortices")
    plt.title("Number of vortices as a function of time for a $128$x$128$ matrix, \n over " + str(R) + "\
    realizations, $C_l = " + str(cl) + "$, $N =" + str(N) + "$, $\lambda = 0$")

    for i in range(1,R+1):
        vortexNum = np.loadtxt("Data_S=128/CL_" + str(cl) + "/VortexNum_CL" + str(cl) + "_N" + str(N) + "_R" + str(i) + ".txt", dtype=float)
        plt.plot(iteration, vortexNum)
        if noise:
            stocVortex += vortexNum[-1]
    if noise:
        stocVortex /= R
        avgEnergy[cl_index] = stocAvg
        avgNumVortices[cl_index] = stocVortex
        avgEnergySquared[cl_index] = stocAvg2
    elif cl_index == 0:
        avgEnergy[0] = energy[-1]
        avgNumVortices[0] = vortexNum[-1]
        avgEnergySquared[0] = energy[-1]**2
```

```python
PlotGraphs(1, 0, 250, 0, False)
```

```python
fig = plt.figure(figsize=(5, 3))
plt.plot(CL, avgEnergy)
plt.title("Average Energy Density as a function of Noise")
plt.xlabel("$C_l$")
plt.ylabel("$<E>/N^2$")
```

```python
# specific heat - discrete derivative
# dE/dT vs T = dE/dC_l vs C_l
dE = [0] * 15
for i in range(0,14):
    dE[i] = avgEnergy[i+1] - avgEnergy[i]

fig = plt.figure(figsize=(5, 3))
plt.plot(CL, dE)
plt.title("Specific Heat as a function of pump power ")
plt.xlabel("$C_l$")
plt.ylabel("$C_V$")
```

```python
fig = plt.figure(figsize=(5, 3))
plt.plot(CL, avgNumVortices)
plt.title("Average Number of Vortices and \n Antivortices as a function of Noise")
plt.xlabel("$C_l$")
plt.ylabel("$Vortices$")
```
