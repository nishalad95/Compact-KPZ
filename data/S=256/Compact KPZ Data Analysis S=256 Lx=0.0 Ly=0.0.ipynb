Calibration of KPZ - set $\lambda_x = 0.0, \lambda_y = 0.0$, Lattice size S = 256

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
avgEnergy = [0] * 15
avgNumVortices = [0] * 15
avgEnergySquared = [0] * 15
# 'temperature', pump power
CL = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0]
print(CL)
```

```python
def PlotGraphs(r, cl, N, cl_index, noise=True):
    R = r
    fig = plt.figure(figsize=(5, 3))
    if noise:
        stocAvg = 0.0
        stocAvg2 = 0.0
        stocVortex = 0.0
    for i in range(1,R+1):
        energy = np.loadtxt("Lx=0.0,Ly=0.0/CL_" + str(cl) + "/Energy_CL" + str(cl) + "_N" + str(N) + "_R" + str(i) + ".txt", dtype=float)
        iteration = np.arange(0, len(energy), 1)
        plt.plot(iteration, energy)
        if noise:
            stocAvg += energy[-1]
            stocAvg2 += energy[-1]**2
    if noise:
        stocAvg2 /= R
        stocAvg /= R
    plt.title("Energy as a function of time for a $256$x$256$ matrix, \n over " + str(R) + "\
    realizations, $C_l =" + str(cl) + "$, $N =" + str(N) + "$, $\lambda_x = 0.0 \lambda_y = 0.0$")
    plt.xlabel("time [T]")
    plt.ylabel("E")
    plt.show()

    fig = plt.figure(figsize=(5, 3))
    plt.xlabel("time [T]")
    plt.ylabel("Number of vortices")
    plt.title("Number of vortices as a function of time for a $256$x$256$ matrix, \n over " + str(R) + "\
    realizations, $C_l = " + str(cl) + "$, $N =" + str(N) + "$, $\lambda_x = 0.0 \lambda_y = 0.0$")

    for i in range(1,R+1):
        vortexNum = np.loadtxt("Lx=0.0,Ly=0.0/CL_" + str(cl) + "/VortexNum_CL" + str(cl) + "_N" + str(N) + "_R" + str(i) + ".txt", dtype=float)
        plt.plot(iteration, vortexNum)
        if noise:
            stocVortex += vortexNum[-1]
    if noise:
        stocVortex /= R
        avgEnergy[cl_index] = stocAvg
        avgNumVortices[cl_index] = stocVortex
        avgEnergySquared[cl_index] = stocAvg2
    elif cl_index == 0:
        avgEnergy[0] = energy[-1]
        avgNumVortices[0] = vortexNum[-1]
        avgEnergySquared[0] = energy[-1]**2
```

```python
fig = plt.figure(figsize=(5, 3))
plt.plot(CL, avgEnergy, 'bo')
plt.title("Average Energy Density as a function of Noise \n for a $256x256$ matrix $\lambda_x = 0.0, \lambda_y = 0.0$")
plt.xlabel("$C_l$")
plt.ylabel("$<E>/N^2$")
np.savetxt("energyDensityVsCL_S=256_Lx=0_Ly=0.txt", avgEnergy, delimiter=' ')
```

```python
# specific heat - discrete derivative
# dE/dT vs T = dE/dC_l vs C_l
dE = [0] * 15
for i in range(0,14):
    dE[i] = (avgEnergy[i+1] - avgEnergy[i])/(CL[i+1] - CL[i])

fig = plt.figure(figsize=(5, 3))
plt.plot(CL, dE, 'bo')
plt.title("Specific Heat as a function of pump power \n for a $256x256$ matrix $\lambda_x = 0.0, \lambda_y = 0.0$")
plt.xlabel("$C_l$")
plt.ylabel("$C_V$")
np.savetxt("specificHeatVsCL_S=256_Lx=0_Ly=0.txt", dE, delimiter=' ')
```

```python
# specific heat - discrete derivative
# variance of energy
Cv = [0] * 15
for i in range(1,15):
    Cv[i] = (avgEnergySquared[i] - avgEnergy[i]**2)/CL[i]**2

fig = plt.figure(figsize=(5, 3))
plt.plot(CL, Cv)
plt.title("Specific Heat as a function of pump power \n for a $256x256$ matrix $\lambda_x = 0.0, \lambda_y = 0.0$")
plt.xlabel("$C_l$")
plt.ylabel("$C_V$")
```

```python
fig = plt.figure(figsize=(5, 3))
plt.plot(CL, avgNumVortices, 'bo')
plt.title("Average Number of Vortices plus Antivortices as a \n"
          "function of Noise for a $256x256$ matrix $\lambda_x = 0.0, \lambda_y = 0.0$")
plt.xlabel("$C_l$")
plt.ylabel("$Vortices$")
np.savetxt("numOfVorticesVsCL_S=256_Lx=0_Ly=0.txt", dE, delimiter=' ')
```
