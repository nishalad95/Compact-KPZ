Calibration of KPZ - set $\lambda_x = -0.25, \lambda_y = 0.75$, Lattice size S = 128

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
I = 27
avgEnergy = [0] * I
avgNumVortices = [0] * I
avgEnergySquared = [0] * I
# 'temperature', pump power
CL = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 2.95, 3.0, 3.05, 3.1, 3.15, 3.2, 3.25, 3.3, 3.35, 3.4, 3.45, 3.5, 3.55, 3.6, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0]
np.savetxt("CL_S=128_Lx=-0.25_Ly=0.75.txt", CL, delimiter=' ')
print(CL)
```

```python
def PlotGraphs(r, cl, N, cl_index, noise=True):
    R = r
    fig = plt.figure(figsize=(6, 5))
    if noise:
        stocAvg = 0.0
        stocAvg2 = 0.0
        stocVortex = 0.0
    for i in range(1,R+1):
        energy = np.loadtxt("Lx=-0.25,Ly=0.75/CL_" + str(cl) + "/Energy_CL" + str(cl) + "_N" + str(N) + "_R" + str(i) + ".txt", dtype=float)
        iteration = np.arange(0, len(energy), 1)
        plt.plot(iteration, energy)
        if noise:
            stocAvg += energy[-1]
            stocAvg2 += energy[-1]**2
    if noise:
        stocAvg2 /= R
        stocAvg /= R
    plt.title("Energy as a function of time for a $128x128$ matrix, \n over " + str(R) + "\
    realizations, $C_l =" + str(cl) + "$, $N =" + str(N) + "$, $\lambda_x = -0.25, \lambda_y = 0.75$")
    plt.xlabel("time [T]")
    plt.ylabel("E")
    fig.savefig("Lx=-0.25,Ly=0.75/Graphs/energyTimeCl=" + str(cl) + ".png", dpi=100)
    plt.show()

    fig = plt.figure(figsize=(6, 5))
    plt.xlabel("time [T]")
    plt.ylabel("Number of vortices")
    plt.title("Number of vortices as a function of time for a $128x128$ matrix, \n over " + str(R) + "\
    realizations, $C_l = " + str(cl) + "$, $N =" + str(N) + "$, $\lambda_x = -0.25, \lambda_y = 0.75$")

    for i in range(1,R+1):
        vortexNum = np.loadtxt("Lx=-0.25,Ly=0.75/CL_" + str(cl) + "/VortexNum_CL" + str(cl) + "_N" + str(N) + "_R" + str(i) + ".txt", dtype=float)
        plt.plot(iteration, vortexNum)
        if noise:
            stocVortex += vortexNum[-1]
    fig.savefig("Lx=-0.25,Ly=0.75/Graphs/vorticesTimeCl=" + str(cl) + ".png", dpi=100)
    if noise:
        stocVortex /= R
        avgEnergy[cl_index] = stocAvg
        avgNumVortices[cl_index] = stocVortex
        avgEnergySquared[cl_index] = stocAvg2
    elif cl_index == 0:
        avgEnergy[0] = energy[-1]
        avgNumVortices[0] = vortexNum[-1]
        avgEnergySquared[0] = energy[-1]**2
```

```python
%%script false 
fig, ax = plt.subplots()
major_ticks = np.arange(0, 7.2, 0.5)
minor_ticks = np.arange(0, 7.0, 0.1)
ax.set_xticks(major_ticks, minor = False)
ax.set_xticks(minor_ticks, minor = True)
#print(len(CL), len(avgEnergy))
ax.plot(CL, avgEnergy, 'b.')
plt.title("Average Energy Density as a function of Noise \n for a 128x128 matrix $\lambda_x = -0.25, \lambda_y = 0.75$")
plt.xlabel("$C_l$")
plt.ylabel("$<E>/S^2$")
np.savetxt("energyDensityVsCL_S=128_Lx=-0.25_Ly=0.75.txt", avgEnergy, delimiter=' ')
```

```python
%%script false 
# specific heat - discrete derivative
# dE/dT vs T = dE/dC_l vs C_l
dE = [0] * I
for i in range(0,I-1):
    dE[i] = (avgEnergy[i+1] - avgEnergy[i])/(CL[i+1] - CL[i])

fig, ax = plt.subplots()
major_ticks = np.arange(0, 7.2, 0.5)
minor_ticks = np.arange(0, 7.0, 0.1)
ax.set_xticks(major_ticks, minor = False)
ax.set_xticks(minor_ticks, minor = True)
ax.set_ylim([0, 3.0])
ax.plot(CL, dE, 'b.')
plt.title("Specific Heat as a function of pump power \n for a 128x128 matrix $\lambda_x = -0.25, \lambda_y = 0.75$")
plt.xlabel("$C_l$")
plt.ylabel("$C_V$")
np.savetxt("specificHeatVsCL_S=128_Lx=-0.25_Ly=0.75.txt", dE, delimiter=' ')
```

```python
%%script false
# specific heat - discrete derivative
# variance of energy
Cv = [0] * I
for i in range(1,I):
    Cv[i] = (avgEnergySquared[i] - avgEnergy[i]**2)/CL[i]**2

fig = plt.figure(figsize=(6, 5))
plt.plot(CL, Cv)
plt.title("Specific Heat as a function of pump power \n for a $128x128$ matrix $\lambda_x = -0.25, \lambda_y = 0.75$")
plt.xlabel("$C_l$")
plt.ylabel("$C_V$")
```

```python
fig, ax = plt.subplots()
major_ticks = np.arange(0, 7.2, 0.5)
minor_ticks = np.arange(0, 7.0, 0.1)
ax.set_xticks(major_ticks, minor = False)
ax.set_xticks(minor_ticks, minor = True)
ax.plot(CL, avgNumVortices, 'b.')
plt.title("Average Number of Vortices plus Antivortices as a \n"
          "function of Noise for a 128x128 matrix $\lambda_x = -0.25, \lambda_y = 0.75$")
plt.xlabel("$C_l$")
plt.ylabel("$Vortices + Antivortices$")
np.savetxt("numOfVorticesVsCL_S=128_Lx=-0.25_Ly=0.75.txt", avgNumVortices, delimiter=' ')
```
