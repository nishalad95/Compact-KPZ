# Compact KPZ - Investigating Nonlinear behaviour

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## S=64

```python
CL1 = np.loadtxt("data/S=64/averagedData/CL_S=64_Lx=0.0_Ly=0.0.txt", dtype=float, delimiter=' ')
CL2 = np.loadtxt("data/S=64/averagedData/CL_S=64_Lx=0.25_Ly=0.75.txt", dtype=float, delimiter=' ')
CL3 = np.loadtxt("data/S=64/averagedData/CL_S=64_Lx=-0.25_Ly=0.75.txt", dtype=float, delimiter=' ')
CL4 = np.loadtxt("data/S=64/averagedData/CL_S=64_Lx=0.5_Ly=0.75.txt", dtype=float, delimiter=' ')
CL5 = np.loadtxt("data/S=64/averagedData/CL_S=64_Lx=-0.5_Ly=0.75.txt", dtype=float, delimiter=' ')

energyDensity1 = np.loadtxt("data/S=64/averagedData/energyDensityVsCL_S=64_Lx=0.0_Ly=0.0.txt", dtype=float, delimiter=' ')
energyDensity2 = np.loadtxt("data/S=64/averagedData/energyDensityVsCL_S=64_Lx=0.25_Ly=0.75.txt", dtype=float, delimiter=' ')
energyDensity3 = np.loadtxt("data/S=64/averagedData/energyDensityVsCL_S=64_Lx=-0.25_Ly=0.75.txt", dtype=float, delimiter=' ')
energyDensity4 = np.loadtxt("data/S=64/averagedData/energyDensityVsCL_S=64_Lx=0.5_Ly=0.75.txt", dtype=float, delimiter=' ')
energyDensity5 = np.loadtxt("data/S=64/averagedData/energyDensityVsCL_S=64_Lx=-0.5_Ly=0.75.txt", dtype=float, delimiter=' ')

fig, ax = plt.subplots()
major_ticks = np.arange(0, 7.2, 0.5)
minor_ticks = np.arange(0, 7.0, 0.1)
ax.set_xticks(major_ticks, minor = False)
ax.set_xticks(minor_ticks, minor = True)

ax.plot(CL1, energyDensity1, 'b.', label='$\lambda_x = 0.0, \lambda_y= 0.0 $')
ax.plot(CL2, energyDensity2, 'r.', label='$\lambda_x = 0.25, \lambda_y= 0.75 $')
ax.plot(CL3, energyDensity3, 'g.', label='$\lambda_x = -0.25, \lambda_y= 0.75 $')
ax.plot(CL4, energyDensity4, 'c.', label='$\lambda_x = 0.5, \lambda_y= 0.75 $')
ax.plot(CL5, energyDensity5, 'm.', label='$\lambda_x = -0.5, \lambda_y= 0.75 $')

plt.xlabel('$C_l$')
plt.ylabel('$<E>/S^2$')
plt.title('Average Energy Density as a function of Noise \n for a 64x64 matrix for varying nonlinear values')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig.savefig("data/S=64/OverlapEnergyDensityCl.png", dpi=100)
```
