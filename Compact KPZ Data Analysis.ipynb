```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
avgEnergy = [0] * 15
avgNumVortices = [0] * 15
```

```python
energy = np.loadtxt("CL_0/Energy_CL0_N250_R1.txt", dtype=float)
iteration = np.arange(0, len(energy), 1)
fig = plt.figure(figsize=(5, 3))
plt.plot(iteration, energy)
plt.title("Energy as a function of time for a \n $64$x$64$ matrix, $C_l = 0.0$, $N = 250$, $\lambda = 0$")
plt.xlabel("time [T]")
plt.ylabel("E")
plt.show()
# only one unique realisation for a particular initial state, due to zero noise

fig = plt.figure(figsize=(5, 3))
plt.xlabel("time [T]")
plt.ylabel("Number of vortices")
plt.title("Number of vortices as a function of time for a \n $64$x$64$ matrix, $C_l = 0.0$, $N = 250$, $\lambda = 0$")
vortexNum = np.loadtxt("CL_0/VortexNum_CL0_N250_R1.txt", dtype=float)
plt.plot(iteration, vortexNum)
avgEnergy[0] = energy[-1]
avgNumVortices[0] = vortexNum[-1]
```

```python
R = 10
fig = plt.figure(figsize=(5, 3))
stocAvg = 0.0
stocVortex = 0.0
for i in range(1,R+1):
    energy = np.loadtxt("CL_0.5/Energy_CL0.5_N250_R" + str(i) + ".txt", dtype=float)
    iteration = np.arange(0, len(energy), 1)
    plt.plot(iteration, energy)
    stocAvg += energy[-1]
stocAvg /= R
plt.title("Energy as a function of time for a $64$x$64$ matrix, \n over 10 \
realizations, $C_l = 0.5$, $N = 250$, $\lambda = 0$")
plt.xlabel("time [T]")
plt.ylabel("E")
plt.show()

fig = plt.figure(figsize=(5, 3))
plt.xlabel("time [T]")
plt.ylabel("Number of vortices")
plt.title("Number of vortices as a function of time for a $64$x$64$ matrix, \n over 10 \
realizations, $C_l = 0.5$, $N = 250$, $\lambda = 0$")

for i in range(1,R):
    vortexNum = np.loadtxt("CL_0.5/VortexNum_CL0.5_N250_R" + str(i) + ".txt", dtype=float)
    plt.plot(iteration, vortexNum)
    stocVortex += vortexNum[-1]
    
stocVortex /= R
avgEnergy[1] = stocAvg
avgNumVortices[1] = stocVortex
```

```python
# specific heat
# dE/dT vs T = dE/dC_l vs C_l
dE = [0] * 10
for i in range(0,10):
  dE[i] = avgEnergy[i+1] - avgEnergy[i]  
dC_l = np.arange(0.25, 5.0, 0.5)
    
fig = plt.figure(figsize=(5, 3))
plt.plot(dC_l, dE)
plt.title("Specific Heat as a function of pump power ")
plt.xlabel("$C_l$")
plt.ylabel("$C_V$")
```
